```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time, glob
```

```python
alltxtFiles=glob.glob('*.txt')

openAllFiles=[]
for files in alltxtFiles:
    openAllFiles+=open(files,"r")
     
diningNames=[]
for dining in openAllFiles:
    diningNames.append(dining.split(',')[1])
```

```python
dinersWithCount={}
for diners in diningNames:
    if diners in dinersWithCount:
        storeValue=dinersWithCount[diners]
        storeValue+=1
        dinersWithCount.update( {diners : storeValue} )
    else:
        dinersWithCount.update( {diners:1} )
        
nameOfDiners=dinersWithCount.keys()
count = list(dinersWithCount.values())
```

```python
plt.figure(figsize=(20, 8))
plt.bar(nameOfDiners,count)
plt.title("Total times Students Visited Each Diner")
plt.ylabel('Count')
plt.xlabel("Name of Diner")
plt.show()
```
